```python
import matplotlib.pyplot as plt
import jax.numpy as np
from jax import grad, jit, vmap
import numpy as npi

N = 500
r = 1/N
a1 = 0
a2 = np.pi*0.55

def get_xy(m, n, a1, a2, r):
    return (r*np.cos(a1)*m + r*np.cos(a2+a1)*n, r*np.sin(a1)*m + r*np.sin(a2+a1)*n)

su = [0.0, 0.0, 0.0]
kth = [0, 0, 97]
uppsala = [c/100*255 for c in [64, 0, 19]]
print(uppsala)
colors = [kth, uppsala, su]
colors = [[c/255 for c in color] for color in colors]
colors = [[0 for c in colors] for color in colors]
h = 10
fig = plt.figure(figsize=(h, h))
for n in range(0, N, 2):
    for m in range(0, N, 2):
        #m = m+n%2
        if npi.random.rand()*0.9 < (n / N):
            x1, y1 = get_xy(m, n, a1, a2, r)
            x2, y2 = get_xy(m+2, n, a1, a2, r)
            x3, y3 = get_xy(m+2, n+2, a1, a2, r)
            xy = np.array([[x1, y1], [x2, y2], [x3, y3]])
            p = plt.Polygon(xy, color=colors[npi.random.randint(0, len(colors))], linewidth=0)
            plt.gca().add_patch(p)
#plt.
plt.gca().set_aspect(0.6)
plt.xlim(0, 0.8)
plt.axis('off')
plt.tight_layout()


fig.savefig("/home/emastr/Desktop/background.pdf", bbox_inches="tight", pad_inches=0)

```
